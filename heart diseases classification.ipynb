6 -Encoding Categorical Features :
Machine learning models cannot process categorical (text) data directly. Therefore, categorical features must be converted into numerical values before training the model. In this project, we use two different encoding techniques depending on the nature of the feature.

a- Label Encoding :
Label Encoding assigns a unique integer to each category. It is suitable for binary categorical features (features with only two categories), such as Sex (Male/Female) and ExerciseAngina (Yes/No). Since these features have only two possible values, representing them as 0 and 1 is sufficient and does not introduce unnecessary complexity.

b- One-Hot Encoding :
One-Hot Encoding creates a separate binary column for each category. It is recommended for features that contain more than two categories, such as ChestPainType, RestingECG, and ST_Slope. This approach prevents the model from assuming an incorrect ordinal relationship between categories.

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
```

7 - We initialize LabelEncoder and OneHotEncoder objects and assign them to shorter variable names (le and ohe) for easier reuse and improved code readability.

note : The handle_unknown='ignore' parameter ensures that the encoder does not raise an error when it encounters unseen categories during prediction. Instead, unknown categories are ignored, making the model more robust when working with new data.

```python
la = LabelEncoder()
ohe = OneHotEncoder(handle_unknown = 'ignore')
```

```python
df['sex'] = la.fit_transform(df['Sex'])
```

9- Apply One-Hot Encoding :
In this step, we apply One-Hot Encoding to the ChestPainType feature. Since this column contains multiple categorical values, each category is converted into a separate binary column (0 or 1).

We apply One-Hot Encoding to the ChestPainType column, transforming each category into a separate binary feature. The toarray() method converts the sparse matrix returned by the encoder into a regular NumPy array, making it easier to work with in the following steps.

15 - Data Splitting :
The dataset was divided into features (X) and target (y), where HeartDisease was used as the target variable.
Then, the data was split into training and testing sets using train_test_split().
shuffle=True was applied to randomly shuffle the data before splitting, reducing the risk of bias from the original data order.
stratify=y was used to preserve the same class distribution of the target variable in both the training and testing sets, which is important for classification tasks.

```python
from sklearn.model_selection import train_test_split
x =df.drop('HeartDisease' , axis = 1)
y =df['HeartDisease']
xtrain , xtest , ytrain , ytest =train_test_split(x , y , shuffle = True , stratify = y)
```

15 - Model initialization :
In this step, multiple machine learning models were imported and initialized, including Logistic Regression, K-Nearest Neighbors (KNN), and Decision Tree.
These models were selected as strong baseline classifiers for the heart disease prediction task.
In the next stage, all models will be trained and evaluated, then their performance will be compared to identify the best-performing model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
lr = LogisticRegression()
kn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
```

16 - Model Training :
The initialized models were trained on the training set (X_train, y_train) so they could learn patterns from the data and make predictions on heart disease cases.
These trained models will then be evaluated and compared in the next step to determine the best performer.

```python
lr.fit(xtrain , ytrain)
kn.fit(xtrain , ytrain)
dt.fit(xtrain , ytrain)
```

Another Method for Model Evaluation : 
In addition to using accuracy scores, another useful way to evaluate a classification model is through the confusion matrix.
While accuracy provides an overall measure of correct predictions, the confusion matrix offers a more detailed view of the model’s performance by showing how predictions are distributed across the actual and predicted classes.
The confusion matrix is composed of four main outcomes:
True Negatives (TN): cases that were correctly predicted as non-heart disease
False Positives (FP): healthy cases that were incorrectly predicted as heart disease
False Negatives (FN): heart disease cases that were incorrectly predicted as healthy
True Positives (TP): cases that were correctly predicted as heart disease
This makes the confusion matrix especially useful for understanding the types of errors made by the model, rather than relying only on a single performance metric such as accuracy

```python
from sklearn.metrics import confusion_matrix
yhatlr = lr.predict(xtest)
```

```python
confusion_matrix(ytest , yhatlr)
```
